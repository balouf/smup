# Using SMUP 

```python
from smup import Smup
matcher = Smup()
matcher.compute()
matcher.display()
```

## Computational parameters

### Norm

```python
matcher.compute(seed=42)
matcher.display()
```

```python
matcher.compute(norm=1, seed=42)
matcher.display()
```

```python
matcher.compute(norm="inf", seed=42)
matcher.display()
```

### Size

```python
matcher.compute(x=100, y=70)
matcher.display()
```

```python
matcher.compute(y=100)
matcher.display()
```

### Number of areas 

```python
matcher.compute(s=5)
matcher.display()
```

```python
matcher.compute(s=50)
matcher.display()
```

## Display parameters

### Colormap

```python
matcher.display(cmap="spring")
```

```python
matcher.display(cmap="autumn")
```

### Centers

```python
matcher.display(draw_centers=False)
```

```python
matcher.display(center_size=50)
```

### Saving

To save the picture, just feed the `save` parameter with filename. Cf https://balouf.github.io/smup/reference/index.html#smup.smup.Smup.display
